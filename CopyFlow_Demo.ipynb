---

## 📖 什么是 CopyFlow？

CopyFlow 是一个开源的内容生产自动化工具，专为解决内容运营的重复劳动而生。

### 核心功能
- ✅ **10 种风格自动生成**：从严肃专业到轻松娱乐，一次生成全部
- ✅ **AI 智能配图**：每套文案自动生成匹配的视觉素材
- ✅ **审核文档整理**：自动生成 Markdown 格式的内容审核文档
- ✅ **对抗式质量保证**：三 Agent 协作，确保内容质量

### 适用场景
- 🎯 多平台内容运营（微信、微博、小红书、抖音等）
- 📱 新媒体矩阵管理
- 📝 内容创作团队效率提升
- 🚀 营销活动快速响应

## 🎬 快速开始

按顺序运行下面的代码块，3 分钟体验完整流程！

## 步骤 1️⃣: 安装 CopyFlow

从 GitHub 克隆仓库并安装依赖（约 1 分钟）

````python
# 克隆仓库
!git clone https://github.com/AIPMAndy/copyflow.git

# 进入目录
%cd copyflow

# 安装依赖
!pip install -q -r requirements.txt

print("\n✅ 安装完成！")
````

## 步骤 2️⃣: 配置 API Keys

**需要准备**：
- 🔑 OpenAI API Key（用于文案生成）→ [获取地址](https://platform.openai.com/api-keys)
- 🖼️ PonyFlash API Key（用于配图生成）→ [获取地址](https://console.ponyflash.com)

**提示**：首次试用可以申请免费额度

````python
import os
from getpass import getpass

# 安全输入 API Keys（不会显示在屏幕上）
print("🔐 请输入你的 API Keys（输入时不会显示）\n")

openai_key = getpass("OpenAI API Key: ")
ponyflash_key = getpass("PonyFlash API Key: ")

# 设置环境变量
os.environ['OPENAI_API_KEY'] = openai_key
os.environ['PONYFLASH_API_KEY'] = ponyflash_key

# 写入 .env 文件
with open('.env', 'w') as f:
    f.write(f"OPENAI_API_KEY={openai_key}\n")
    f.write(f"PONYFLASH_API_KEY={ponyflash_key}\n")

print("\n✅ API Keys 配置完成！")
````

## 步骤 3️⃣: 运行内容生成

现在开始生成内容！你可以修改 `topic` 和 `count` 参数：
- `topic`: 内容主题
- `count`: 生成数量（建议 3-5 套快速体验，完整版 10 套）

````python
# 运行生成（这里生成 5 套内容）
!python copyflow.py run --topic "AI 改变内容创作" --count 5

print("\n✅ 生成完成！")
print("📁 结果保存在 output/ 目录")
````

## 步骤 4️⃣: 查看生成结果

### 4.1 查看目录结构

````python
# 查看生成的文件
!ls -lh output/

print("\n📂 文件说明：")
print("  - style_*.txt: 不同风格的文案")
print("  - image_*.png: AI 生成的配图")
print("  - content-review.md: 审核文档（所有内容汇总）")
````

### 4.2 预览审核文档

这个 Markdown 文档包含了所有生成的文案和配图，方便团队审核

````python
from IPython.display import Markdown, display

# 读取并显示审核文档
with open('output/content-review.md', 'r', encoding='utf-8') as f:
    content = f.read()

display(Markdown(content))
````

### 4.3 预览单个风格的文案

````python
import glob

# 列出所有文案文件
text_files = sorted(glob.glob('output/style_*.txt'))

# 显示第一个风格的文案
if text_files:
    print(f"📝 预览: {text_files[0]}\n")
    print("-" * 50)
    with open(text_files[0], 'r', encoding='utf-8') as f:
        print(f.read())
    print("-" * 50)
    
    print(f"\n💡 提示：共生成 {len(text_files)} 种风格，可修改上面的索引查看其他风格")
else:
    print("❌ 未找到生成的文案文件")
````

### 4.4 预览生成的配图

````python
from IPython.display import Image, display
import glob

# 列出所有图片文件
image_files = sorted(glob.glob('output/image_*.png'))

if image_files:
    print(f"🖼️  共生成 {len(image_files)} 张配图\n")
    
    # 显示前 3 张图片
    for img_path in image_files[:3]:
        print(f"📷 {img_path}")
        display(Image(filename=img_path, width=400))
        print("\n")
    
    if len(image_files) > 3:
        print(f"💡 还有 {len(image_files) - 3} 张图片，可在 output/ 目录查看")
else:
    print("❌ 未找到生成的图片文件")
````

## 步骤 5️⃣: 下载结果

将生成的内容打包下载到本地

````python
import shutil
from datetime import datetime

# 创建压缩包
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
archive_name = f'copyflow_output_{timestamp}'

shutil.make_archive(archive_name, 'zip', 'output')

print(f"✅ 已创建压缩包: {archive_name}.zip")
print("\n📥 下载方式：")
print("  1. 在左侧文件列表中找到压缩包")
print("  2. 右键点击 → Download")
print("\n或者运行下面的代码直接下载：")
````

````python
from google.colab import files

# 直接下载（仅在 Colab 环境有效）
try:
    files.download(f'{archive_name}.zip')
    print("✅ 下载已开始！")
except:
    print("⚠️  请手动从文件列表下载")
````
